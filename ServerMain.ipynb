```python
from pynq import Overlay
from pynq.lib.video import *
from ServerOnPYNQ import *
## 加载base
base = Overlay('base.bit')
```

```python
### 初始化一个hdmi的实例
hdmi_in = base.video.hdmi_in
```

```python
# 进行配置，默认是24bit的BGR


hdmi_in.configure()
hdmi_in.mode.width = 1920
hdmi_in.mode.height = 1080
print(hdmi_in.mode.width,hdmi_in.mode.height)

colorspace_in = base.video.hdmi_in.color_convert
bgr2rgb = [0, 0, 1,
           0, 1, 0, 
           1, 0, 0,
           0, 0, 0]

colorspace_in.colorspace = bgr2rgb
colorspace_in.colorspace
```

```python
# 执行
hdmi_in.start()
```
